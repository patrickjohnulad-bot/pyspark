````python
import pandas as pd
````

````python
pd.DataFrame(np.array([age, gender]).T,
             columns = ['Age', 'Gender'])
````

````python
data_sim.head()
````

````python
data_sim['Gender'].value_counts() #.plot(kind="pie")
````

````python
data_sim.groupby(['Gender'])['Age'].mean()#.plot(kind="pie")
````
